# DataFrame

API: [pandas.DataFrame](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.html)

## Import data

### Import data from dictionaries

```python
import pandas as pd
```

```python
data = {'col_1': [3, 2, 1, 0], 'col_2': ['a', 'b', 'c', 'd']}
pd.DataFrame.from_dict(data=data, orient='columns')
```

```python
data = {'row_1': [3, 2, 1, 0], 'row_2': ['a', 'b', 'c', 'd']}
pd.DataFrame.from_dict(data, orient='index')
```

```python
pd.DataFrame.from_dict(data, orient='index', columns=['A', 'B', 'C', 'D'])
```

### Import data from items

Deprecated since version 0.23.0: from_items is deprecated and will be removed in a future version. Use `DataFrame.from_dict(dict(items))` instead. `DataFrame.from_dict(OrderedDict(items))` may be used to preserve the key order.

```python
items = dict([('row_1', [3, 2, 1, 0]), ('row_2', ['a', 'b', 'c', 'd'])])
pd.DataFrame.from_dict(dict(items), columns=None, orient='index')
```

```python
pd.DataFrame.from_dict(dict(items), orient='index', columns=['A', 'B', 'C', 'D'])
```

### Area Plots

API: [pandas.DataFrame](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.plot.area.html#pandas.DataFrame.plot.area)

```python
df = pd.DataFrame({
    'sales': [3, 2, 3, 9, 10, 6],
    'signups': [5, 5, 6, 12, 14, 13],
    'visits': [20, 42, 28, 62, 81, 50],
}, index=pd.date_range(start='2018/01/01',
                       end='2018/07/01',
                       freq='M'))
ax = df.plot.area()
```

```python
ax = df.plot.area(stacked=False)
```

```python
ax = df.plot.area(y='sales')
```

```python
df = pd.DataFrame({
    'sales': [3, 2, 3],
    'visits': [20, 42, 28],
    'day': [1, 2, 3],
})
ax = df.plot.area(x='day')
```

### Bar Plots

API: [DataFrame.plot.bar](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.plot.bar.html#pandas.DataFrame.plot.bar)

```python
df = pd.DataFrame({'lab':['A', 'B', 'C'], 'val':[10, 30, 20]})
ax = df.plot.bar(x='lab', y='val', rot=0)
```

```python
speed = [0.1, 17.5, 40, 48, 52, 69, 88]
lifespan = [2, 8, 70, 1.5, 25, 12, 28]
index = ['snail', 'pig', 'elephant',
         'rabbit', 'giraffe', 'coyote', 'horse']
df = pd.DataFrame({'speed': speed,
                   'lifespan': lifespan}, index=index)
ax = df.plot.bar(rot=0)
```

```python
axes = df.plot.bar(rot=0, subplots=True)
axes[1].legend(loc=2)  # doctest: +SKIP
```

```python
ax = df.plot.bar(y='speed', rot=0)
```

```python
ax = df.plot.bar(x='lifespan', rot=0)
```

### Horizontal Bar Plots

API: [pandas.DataFrame.plot.barh](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.plot.barh.html#pandas.DataFrame.plot.barh)

```python
df = pd.DataFrame({'lab':['A', 'B', 'C'], 'val':[10, 30, 20]})
ax = df.plot.barh(x='lab', y='val')
```

```python
speed = [0.1, 17.5, 40, 48, 52, 69, 88]
lifespan = [2, 8, 70, 1.5, 25, 12, 28]
index = ['snail', 'pig', 'elephant',
         'rabbit', 'giraffe', 'coyote', 'horse']
df = pd.DataFrame({'speed': speed,
                   'lifespan': lifespan}, index=index)
ax = df.plot.barh()
```

```python
ax = df.plot.barh(y='speed')
```

```python
ax = df.plot.barh(x='lifespan')
```

## Density Plots

Requires to install `scipy` manually.

API: [pandas.DataFrame.plot.density](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.plot.density.html#pandas.DataFrame.plot.density)

```python
df = pd.DataFrame({
    'x': [1, 2, 2.5, 3, 3.5, 4, 5],
    'y': [4, 4, 4.5, 5, 5.5, 6, 6],
})
df.plot()
```

```python
ax = df.plot.kde()
```

```python
ax = df.plot.kde(bw_method=0.3)
```

```python
ax = df.plot.kde(bw_method=3)
```

```python
ax = df.plot.kde(ind=[1, 2, 3, 4, 5, 6])
```
